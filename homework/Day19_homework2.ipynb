```python
import pandas as pd
import numpy as np
```

```python
#1.畫出箱型圖，並判斷哪個欄位的中位數在300~400之間?
boston = pd.read_csv('boston.csv')
boston
```

```python
# 'TAX' 和 'B' 這兩欄的中位數在 300 ~ 400 之間
import matplotlib.pyplot as plt 
plt.figure(figsize=(14, 10))
boston.boxplot()
plt.axhline(y=300, color='r')
plt.axhline(y=400, color='r')
```

```python
#2. 畫出散佈圖 x='NOX', y='DIS' ，並說明這兩欄位有什麼關係?
plt.figure(figsize=(14, 10))
plt.scatter(x=boston['NOX'], y=boston['DIS'])
plt.show()
```

```python
corr = pd.DataFrame(boston.corr())
corr
```

```python
# 'NOX' 和 'DIS' 分布呈現高度負相關
corr.loc['NOX', 'DIS']
```
